```python
import os
import dlt
from dotenv import load_dotenv
import requests
import time
```

```python
# get the secret key
load_dotenv()
secret_key = os.getenv(key="WEATHER_STACK_API")
```

```python
from requests.exceptions import HTTPError

# get data from api for the below cities
cities = ["Bangalore", "New Delhi", "Hyderabad", "Pune", "Chennai", "Kolkata"]

def extract_weather_data(city_list):
    for city in city_list:

        # delay to avoid maximum rate limitation of API call
        time.sleep(1.5)

        try:
            base_url = f"https://api.weatherstack.com/current?access_key={secret_key}"
            querystring = {"query": f"{city}"}
            response = requests.get(url=base_url, params=querystring)

            response.raise_for_status()

            yield response.json()
        except HTTPError as e:
            yield {"Error": str(e), "city": city}
        except Exception as e:
            yield {"Error" : str(e), "city": city}
        
```

```python
weather_generator = extract_weather_data(cities)

for weather_data in weather_generator:
    print(weather_data["current"]["air_quality"]["us-epa-index"])

```
